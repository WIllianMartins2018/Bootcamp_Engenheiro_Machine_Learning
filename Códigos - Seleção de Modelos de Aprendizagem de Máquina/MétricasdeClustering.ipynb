```python
from sklearn.model_selection import train_test_split

#Divindo em treino e teste as duas bases, binarizada e n√£o binzarizada
train_data, test_data, train_labels, test_labels = train_test_split(df, labels, test_size= 0.25, random_state = 42)

train_data1, test_data1, train_labels1, test_labels1 = train_test_split(df1, labels1, test_size = 0.25, random_state = 42)
```

```python
#BASELINE: COMPARANDO COM UM MODELO ALEATORIO
baseline_preds = np.random.choice([0,1,2], size = len(test_labels))
print(baseline_preds)
```

```python
from sklearn import metrics 
from sklearn.metrics import cluster

#Avaliando o Baseline

#Ponto de Vista N√£o Supervisionado
print('Coeficiente de Silhueta\n', metrics.silhouette_score(test_data, baseline_preds))
print('\nDavies-Bouldin Score\n', metrics.davies_bouldin_score(test_data, baseline_preds))

#Ponto de Vista Supervisionado
print('\nMatriz de Contig√™ncia\n', metrics.cluster.contingency_matrix(test_labels, baseline_preds))
print('\nMutual Information\n', metrics.mutual_info_score(test_labels, baseline_preds))
```

```python
#K-MEANS
from sklearn.cluster import KMeans

clustering = KMeans(n_clusters=3, random_state=42)
#treinando o Modelo com a base de dados n√£o binzarizado
clustering.fit(train_data)
```

```python
#Predi√ß√µes
predicition1_labels = clustering.predict(test_data)

#Exibindo dataframe com real e previto
p = pd.DataFrame({'Real': test_labels, 'Previsto': predicition1_labels})
p.head(10)
```

```python
#Avaliando o Modelo K-means, com a base de dados n√£o binarizada

#Ponto de Vista N√£o Supervisionado
print('Coeficiente de Silhueta\n', metrics.silhouette_score(test_data, predicition1_labels))
print('\nDavies-Bouldin Score\n', metrics.davies_bouldin_score(test_data, predicition1_labels))

#Ponto de Vista Supervisionado
print('\nMatriz de Contig√™ncia\n', metrics.cluster.contingency_matrix(test_labels, predicition1_labels))
print('\nMutual Information\n', metrics.mutual_info_score(test_labels, predicition1_labels))
```
