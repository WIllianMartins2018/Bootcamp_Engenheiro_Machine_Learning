```python
#Importando as bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
#Gerando os parâmetros de uma séries sintética
#Seguindo com análise de decomposição

#Numero de amostras
n = 100
x = np.linspace(0, 2*np.pi, n)
#Frequência de amostragem da série
fs = 1/(x[1] - x[0])
#Frequência da Série
f = fs/20
```

```python
#Gerador de Série
def generate_series(x, freq, opt='linear_trend+seasonal'):
    if opt == 'linear_trend':
        fx = x
    elif opt == 'linear_trend+seasonal':
        fx = x + np.sin(2*np.pi*freq*x)
    elif opt == 'seasonal':
        fx = np.sin(2*np.pi*freq*x)
    else:
        raise ValueError("Unkown opt parameter.")
    return fx
```

```python
#Método para plot da decomposição
def plot_decomposition(decomposition):
     
    # plot handle
    fig, ax = plt.subplots(figsize=(6, 6), nrows=4)
    
    # original signal
    ax[0].plot(decomposition.observed)
    ax[0].set_title("Series")
    
    # trend component
    ax[1].plot(decomposition.trend)
    ax[1].set_title("Trend")
    
    # seasonal component
    ax[2].plot(decomposition.seasonal)
    ax[2].set_title("Seasonal")
    
    # residue
    ax[3].plot(decomposition.resid)
    ax[3].set_title("Residue")
    
    # plot configuration
    plt.tight_layout()
    plt.show()
```

```python
#Números de Pontos por Período
n_points_in_period = int(fs/f)
print(f"Número de Pontos por Período: {n_points_in_period}")
```

```python
#Série: Tendência Linear
series = generate_series(x, f, opt='linear_trend')
#model, Modelo aditivo ou multiplicativo
decomposition = seasonal_decompose(series, model='additive', period=n_points_in_period)
plot_decomposition(decomposition)
```

```python
#Série: tendencia + sazonalidade
series = generate_series(x, f, opt='linear_trend+seasonal')
#model, Modelo aditivo ou multiplicativo
decomposition = seasonal_decompose(series, model='additive', period=n_points_in_period)
plot_decomposition(decomposition)
```

```python
#Série: sazonalidade
series = generate_series(x, f, opt='seasonal')
#model, Modelo aditivo ou multiplicativo
decomposition = seasonal_decompose(series, model='additive', period=n_points_in_period)
plot_decomposition(decomposition)
```

```python
#Analisando Seríes Temporais com Dados Reais
df = pd.read_csv("https://pycourse.s3.amazonaws.com/monthly_air_passengers.csv", index_col='Month')
df.head()
```

```python
#plot
df.plot(figsize=(10,5))
```

```python
#Decomposição
decomposition = seasonal_decompose(df.values, model='additive', period=12)
plot_decomposition(decomposition)
```

```python
#Verificação de Estacionáridade
#teste de Hipótese
result = adfuller(df)
print(f'p-value: {result[1]}')
```

```python
#Diferenciando a Série (Estatégia para torna a série estacionária)
df_diff = df.diff().dropna()
```

```python
#teste da série diferenciada
result = adfuller(df_diff.dropna())
print(f'p-value: {result[1]}')
```

```python
#Visualização da ACF e PACF (Só é válida quando tempos uma série estacionária)
fig, ax = plt.subplots(ncols=2, figsize=(15,7))
plot_acf(df_diff, ax = ax[0])
plot_pacf(df_diff, ax = ax[1])
plt.show()
```

```python
#MODELOS ARIMA
#SELEÇÃO DE PADRÃO EM MODELOS ARIMA
import pmdarima as pm
```

```python
#Divisão de treino e teste
n, p = len(df), 0.7
train = df[:int(n*p)]
test = df[int(n*p):]
```

```python
#treinando o Modelo
model = pm.auto_arima(train, seasonal=True, m=12, trace=True, erro_action='ignore', suppress_warnings=True)

#Relatório
print(model.summary())
```

```python
#Predição
forecast = model.predict(n_periods=len(test))
```

```python
#Plot Predição
plt.figure(figsize=(10,5))
plt.plot(np.arange(int(n*p)), train.values, '.-', label='Treino')
plt.plot(np.arange(int(n*p), n), test.values, '.-', label='Teste')
plt.plot(np.arange(int(n*p), n), forecast, '.-', label='Predição')
plt.grid()
plt.legend()
plt.show()
```

```python
#RMSE
from sklearn.metrics import mean_squared_error

rmse = np.sqrt(mean_squared_error(test,forecast))
print(f'RMSE do modelo no conjunto de teste: {rmse}')
```
